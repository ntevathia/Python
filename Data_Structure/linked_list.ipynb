```python
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None

    def get_data(self):
        return self.data

    def get_next(self):
        return self.next

    def set_data(self, value):
        self.value = value

    def set_next(self, Node):
        self.next = Node
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
    
    def is_empty(self):
        return self.head == None
    
    def add(self, value):
        p = Node(value)
        if self.is_empty():
            self.head = p
        else:
            tmp = self.head
            while tmp.next != None:
                tmp = tmp.next    
            tmp.next = p
            
    def size(self):
        tmp = self.head
        if self.is_empty():
            print("Empty List!!")
            return
        count = 0
        while tmp!=None:
            tmp = tmp.next
            count += 1
        return count
    
    def search(self, value):
        if self.is_empty():
            print("Empty List!!")
            return

        tmp = self.head
        if tmp.data == value:
            return True
        else:
            while tmp.next!=None and tmp.data!=value:
                tmp = tmp.next
            if tmp.next==None:
                return False
            else:
                return True
                
    
    def remove(self, value):
        if self.is_empty():
            print("Empty List!!")
            return
        
        # if value is at first position
        if self.head.data == value:
            self.head = self.head.next
            print("Removed")
            return True
         
        cur = self.head
        while cur.next!=None and cur.data!=value:
            prev = cur
            cur = cur.next

        if cur.next == None:
            print("value not present in the list")
            return False # value not found
        else:
            prev.next = cur.next
            print("Removed")
            return True
        
    def __str__(self):
        if self.is_empty():
            print("Empty List!!")
            return 
        else:
            cur = self.head
            l = []
            while cur.next != None:
                l.append(cur.data)
                cur = cur.next
            return str(l)
```

## Insert at specific position

```python
class LinkedList:
    def __init__(self):
        self.head = None
    
    def is_empty(self):
        return self.head == None
    
    def add(self, value):
        p = Node(value)
        if self.is_empty():
            self.head = p
        else:
            tmp = self.head
            while tmp.next != None:
                tmp = tmp.next    
            tmp.next = p
    
    def insert(self, value, pos):
        if self.is_empty():
            print("Empty List!!")
            return
        
        p = Node(value)
        
        if pos == 1:
            p.next = self.head
            self.head = p
            print("Inserted at first position")
            return
        
        count = 0
        cur = self.head
        while cur.next!=None and count!=pos-1:
            prev = cur
            cur = cur.next
            count += 1
        
        # handles last position 
        print("count",count)
        if cur.next==None and count+1==pos:
            cur.next = p
            print("Inserted at last position!!")
            return
        elif cur.next==None:
            print("Position not found!")
            return
        else:
            prev.next = p
            p.next = cur
            print("Inserted!!")
        
```
